```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from performance_statistics import load_overall_statistics_x_y
overall_statistics,X,y = load_overall_statistics_x_y()
from sklearn.decomposition import PCA
```

```python
for column in X.columns:
    plt.scatter(X[column].values,y, alpha  = 0.3)
    plt.xlabel(column)
    #plt.ylabel(Salary)
    plt.show()
```

```python
pca = PCA()
x_pca = pca.fit_transform(X.values)
for columnname, column in zip(X.columns,x_pca.T):
    plt.scatter(column,y, alpha  = 0.3)
    plt.xlabel(columnname)
    plt.show()
```
